# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
def One_dim(i_step):
    step_x = np.random.random(i_step)-1/2
    
    return (np.cumsum(step_x))

i_step = 1
i_walk = 55

x_f = np.zeros(i_walk)
for run in range(i_walk):
    x = One_dim(i_step)
    x_f[run] = x[-1]
    
pylab.plot(x_f,) 
pylab.show() 
```

```python
arr = np.array(x_f)

sub_arr = np.abs(arr[:,None] - arr)

pylab.hist(sub_arr[0],bins=666) 
pylab.show() 
```

```python
import pylab
import random 

i = 777

x = np.zeros(i) 

for n in range(1, i): 
	val = random.randint(1, 2) 
	if val == 1: 
		x[n] = x[n - 1] + 1
	elif val == 2: 
		x[n] = x[n - 1] - 1

pylab.plot(x) 
pylab.show() 
```

```python
arr = np.array(x)

sub_arr = np.abs(arr[:,None] - arr)

pylab.hist(sub_arr[0],bins=88) 
pylab.show() 
```
