# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
#define site(x) and site(y) with with uniform distribution between  [1/2,+1/2] 
def path(walk):
    
    site_x = np.array([0])
    site_y = np.array([0])
    
    for i in range(len(walk[0])):
        site_x = np.append(site_x, site_x[i] + walk[0][i])
        site_y = np.append(site_y, site_y[i] + walk[1][i])

    return [site_x, site_y]

P_x = np.random.random((5000,)) - 0.5
P_y = np.random.random((5000,)) - 0.5

site = path([P_x, P_y])

plt.plot(site[0], site[1])
```

```python
# figure of site of x, y, and r 
plt.figure(figsize=(25, 6))

r = np.sqrt(site[1]**2 + site[1]**2)

plt.subplot(131)
pylab.plot(site[0]) 


plt.subplot(132)
pylab.plot(site[1]) 


plt.subplot(133)
pylab.plot(r) 
pylab.show() 
```

```python

def r(site):
    output = site[0] * site[1]
    return output

def continue_case(steps, n):
    output_x = np.zeros((n,))
    output_y = np.zeros((n,))
    
    for i in range(n):
        P_x = np.random.random((steps,)) - 0.5
        P_y = np.random.random((steps,)) - 0.5
        terminal_site_x = np.sum(P_x)
        terminal_site_y = np.sum(P_y)
        output_x[i] = terminal_site_x
        output_y[i] = terminal_site_y
    return [output_x, output_y]


def randomstep_2d(n_steps, step_size, n):
    
    output_x = np.zeros((n,))
    output_y = np.zeros((n,))
    
    for i in range(n):
        P_x = np.random.randint(2, size=n_steps)
        P_y = np.random.randint(2, size=n_steps)
        P_x = step_size * (P_x - 0.5)
        P_y = step_size * (P_y - 0.5)
        output_x[i] = np.sum(P_x)
        output_y[i] = np.sum(P_y)
    
    output = np.array([output_x, output_y])
    return output


output_1 = randomstep_2d(10000, 0.5, 10000)
plt.hist(r(output_1), bins=50)
plt.show()
```

```python
def r(site):
    output = site[0] * site[1]
    return output

def random_angle(steps, n):
    output_x = np.zeros((n,))
    output_y = np.zeros((n,))

    for i in range(n):
        P = np.random.random((steps,))
        P = P * 2 * np.pi
        x = np.cos(P)
        y = np.sin(P)
        output_x[i] = np.sum(x)
        output_y[i] = np.sum(y)

    output = np.array([output_x, output_y])
    return output


output3 = random_angle(1000, 1000)

plt.hist(r(output3), bins=100)
plt.show()
```
