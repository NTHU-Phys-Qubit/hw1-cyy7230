# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
def Two_dim(n_step):
    step_x = np.random.random(i_step)-1/2
    step_y = np.random.random(i_step)-1/2
    return (np.cumsum(step_x),np.cumsum(step_y))


i_step = 1
i_walk = 87

x_f = np.zeros(i_walk)
for run in range(i_walk):
    x = Two_dim(i_step)
    x_f[run] = x[-1]
    
y_f = np.zeros(i_walk)
for run in range(i_walk):
    y = Two_dim(i_step)
    y_f[run] = y[-1]

r_f = np.sqrt(x_f**2 + y_f**2)

pylab.plot(x_f) 
pylab.show() 

pylab.plot(y_f) 
pylab.show() 

pylab.plot(r_f) 
pylab.show() 
```

```python
pylab.plot(x_f,y_f,"o") 
pylab.show() 
```

```python
r_f = np.sqrt(x_f**2 + y_f**2)

arr = np.array(r_f)

sub_arr = np.abs(arr[:,None] - arr)


pylab.hist(sub_arr[0],bins=100)  
pylab.show() 
```

```python
import pylab
import random

i = 777

x = np.zeros(i) 
y = np.zeros(i) 


for n in range(1, i): 
	val = random.randint(1, 4) 
	if val == 1: 
		x[n] = x[n - 1] + 1
		y[n] = y[n - 1] 
	elif val == 2: 
		x[n] = x[n - 1] - 1
		y[n] = y[n - 1] 
	elif val == 3: 
		x[n] = x[n - 1] 
		y[n] = y[n - 1] + 1
	else: 
		x[n] = x[n - 1] 
		y[n] = y[n - 1] - 1
       
r = np.sqrt(x**2 + y**2)
    
pylab.plot(x) 
pylab.show() 

pylab.plot(y) 
pylab.show() 

pylab.plot(r) 
pylab.show() 
```

```python
r = np.sqrt(x**2 + y**2)

arr = np.array(r)

sub_arr = np.abs(arr[:,None] - arr)


pylab.hist(sub_arr[0],bins=100)  
pylab.show() 
```

```python
def Pol_cor(i_step):
    step_theta = np.random.random(i_step) * 2 *np.pi
        
    return (np.cumsum(step_theta))

i_step = 1 
i_walk = 999

theta_f = np.zeros(i_walk)
for run in range(i_walk):
    theta = Pol_cor(i_step)
    theta_f[run] = theta[-1]
    
pylab.plot(np.cos(theta_f)) 
pylab.show() 

pylab.plot(np.sin(theta_f)) 
pylab.show() 

r = np.sqrt(np.cos(theta_f)**2 + np.sin(theta_f)**2)

pylab.plot(r) 
pylab.show() 
    
```
